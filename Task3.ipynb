```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_webpage(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract text (paragraphs in this example)
        paragraphs = [p.text for p in soup.find_all('p')]

        # Extract links
        links = [a['href'] for a in soup.find_all('a', href=True)]

        # Extract image sources
        images = [img['src'] for img in soup.find_all('img', src=True)]

        return {
            'text': paragraphs,
            'links': links,
            'images': images
        }
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Type', 'Content'])

        for key, values in data.items():
            for value in values:
                writer.writerow([key, value])

# Example usage
url = 'https://notes.ayushsharma.in/technology'  # Replace with the URL you want to scrape
scraped_data = scrape_webpage(url)

if scraped_data:
    save_to_csv(scraped_data, 'scraped_data.csv')
    print("Data has been scraped and saved to 'scraped_data.csv'")
```

```python
import csv
from collections import defaultdict

def read_csv_file(filename):
    data = defaultdict(list)

    with open(filename, 'r', newline='', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row

        for row in csv_reader:
            if len(row) == 2:
                data_type, content = row
                data[data_type].append(content)

    return data

def display_csv_contents(data):
    print("Contents of the CSV file:")
    print("-------------------------")

    for data_type, contents in data.items():
        print(f"\n{data_type.capitalize()}:")
        for i, content in enumerate(contents, 1):
            print(f"  {i}. {content}")

# Example usage
filename = 'scraped_data.csv'  # Make sure this matches the filename used in the web scraper script
csv_data = read_csv_file(filename)
display_csv_contents(csv_data)
```
